```python
import pandas as pd

df = pd.read_csv(datasheet)
df.head(5)
```

```python
df.describe()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
df.head(2)
```

```python
df.columns
```

**PREPROCESSING**

```python
Features = ['dt', 'dur', 'tot_dur', 'pktrate', 'port_no', 'rx_kbps', 'tot_kbps']
X = df[Features]
```

```python
# Menghapus nilai negatif pada kolom dt, dur, tot_dur

kolom_negatif = ['dt','dur','tot_dur']
X = X[(X[kolom_negatif] >= 0 ).all(axis=1)]
```

```python
df = df.loc[X.index]
```

```python
X.describe()
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**TRAINING**

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=2,
    init='k-means++',
    n_init=50,
    max_iter=500,
    tol=1e-5,
    algorithm='elkan',
    random_state=42,

)
```

```python
cluster = kmeans.fit_predict(X_scaled)
```

```python
cluster_labels = kmeans.labels_
```

```python
df_cluster = X.copy()
df_cluster['cluster'] = cluster_labels
```

**EVALUATE**

```python
# Membuat sample 20000 hasil training dari 5 juta
import numpy as np

sample_size = 200000
idx = np.random.choice(len(X_scaled), size=sample_size, replace=False)
```

```python
X_sample = X_scaled[idx].astype(np.float32)
labels_sample = cluster_labels[idx]
```

```python
from sklearn.metrics import silhouette_score

sil = silhouette_score(X_sample, labels_sample)
print("Silhoutte Score :", sil)
```

```python
df_cluster.groupby('cluster').mean()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=cluster_labels)
plt.title('PCA 2D Clustering')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

pca = PCA(n_components=3)
X_pca3 = pca.fit_transform(X_scaled)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca3[:,0], X_pca3[:,1], X_pca3[:,2], c=cluster_labels, s=2)
plt.title("PCA 3D Clustering")
plt.show()
```

```python
import seaborn as sns

df_cluster = pd.DataFrame(X, columns=Features)
df_cluster['cluster'] = cluster_labels

plt.figure(figsize=(7,5))
sns.boxplot(data=df_cluster, x='cluster', y='pktrate')
plt.show()
```

```python
df_cluster.groupby('cluster').size()
```

```python
counts = df_cluster['cluster'].value_counts().sort_index()

plt.figure(figsize=(6,4))
plt.bar(counts.index.astype(str), counts.values)
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.title('Jumlah Data Per Cluster')
plt.show()
```

```python
from sklearn.metrics import pairwise_distances_argmin_min

closest, distances = pairwise_distances_argmin_min(kmeans.cluster_centers_, X_scaled)
distances[:10]
```
