```python
import pandas as pd

df = pd.read_csv(datasheet)
df.head(5)
```

```python
df.describe()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
df.head(2)
```

```python
df.columns
```

**PREPROCESSING**

```python
Features = ['dt', 'dur', 'tot_dur', 'pktrate', 'port_no', 'rx_kbps', 'tot_kbps']
X = df[Features]
```

```python
# Menghapus nilai negatif pada kolom dt, dur, tot_dur

kolom_negatif = ['dt','dur','tot_dur']
X = X[(X[kolom_negatif] >= 0 ).all(axis=1)]
X = X.reset_index(drop=True)
```

```python
# Menyamakan index df dengan index X

df = df.loc[X.index]
df = df.reset_index(drop=True)
```

```python
X.describe()
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans

kmeans_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('kmeans', KMeans(
        n_clusters=2,
        init='k-means++',
        n_init=50,
        max_iter=500,
        tol=1e-5,
        algorithm='elkan',
        random_state=42
    ))
])
```

**TRAINING**

```python
kmeans_pipeline.fit(X)
```

```python
# Ambil Cluster

cluster_labels = kmeans_pipeline.named_steps['kmeans'].labels_
df['cluster'] = cluster_labels
```

```python
X_cluster = X.copy()
X_cluster['cluster'] = cluster_labels
```

```python
df.head(5)
```

```python
X_cluster.head(5)
```

**EVALUATE**

```python
# Mengambil X yang sudah distandarisasi di pipeline

X_scaler = kmeans_pipeline.named_steps['scaler'].transform(X)
```

```python
# Membuat sample 20000 hasil training dari 5 juta
import numpy as np

sample_size = 200000
idx = np.random.choice(len(X_scaler), size=sample_size, replace=False)
```

```python
X_sample = X_scaler[idx].astype(np.float32)
labels_sample = cluster_labels[idx]
```

```python
# Evaluasi menggunakan silhouette
from sklearn.metrics import silhouette_score

sil = silhouette_score(X_sample, labels_sample)
print("Silhoutte Score :", sil)
```

```python
# Evaluasi menggunakan Davies-Bouldin Index (DBI)
from sklearn.metrics import davies_bouldin_score

dbi = davies_bouldin_score(X_sample, labels_sample)
print("Davies_Bouldin_Score:", dbi)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaler)

plt.figure(figsize=(8,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=cluster_labels)
plt.title('PCA 2D Clustering')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

pca = PCA(n_components=3)
X_pca3 = pca.fit_transform(X_scaler)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca3[:,0], X_pca3[:,1], X_pca3[:,2], c=cluster_labels, s=2)
plt.title("PCA 3D Clustering")
plt.show()
```
